가 5 와 6 모두 나누어지는지, 5 또는 6 으로 나누어지는지, 혹은 두 정수 모

두로는 나누어지지 않지만 둘 중에 하나로만 나누어지는지를 검사하는 프로

그램을 작성하시오. 

하나의 정수를 입력하세요: 10 (enter)

10 은/는 5 와 6 으로 나누어집니까? False

10 은/는 5 혹은 6 으로 나누어집니까? True

10 은/는 5 혹은 6 으로 나누어지지만, 둘 모두로는 나누어지지 않습니까? True.

혹은 두 정수 모두로는 나누어지지 않지만 둘 중에 하나로만 나누어지는지를 검사하기 위해서 하나의

정수를 입력받고 첫 번째 경우 둘다 나누어지는지 확인하기 위해 if문에 조건을 논리곱을 이용하여

둘다 나누어 떨어지는지를 알아봤습니다. bool을 사용하여 true 인지 false인지를 출력하도록 하였습니다.

두 번째 경우 둘 중에 하나로 나누어지는지 확인하기 위해 if문에 조건을 논리합을 이용하여 둘 중 하나

라도 만족을 하면 true 아니면 false가 되도록 설정하였습니다. 세 번째 경우 둘 중 하나로는 나누어 

지는데 둘 다로는 나누어지지 않는지 확인하기 위해 논리합, 논리곱을 같이 사용하여 둘 중 하나가 나눠지고 

둘 다 나눠지면 true, 둘 중 하나가 나눠지고 둘다 나뉘지 않으면 false가 되도록 설정하여 해결하였습니다.

문제 3 번. 도시의 인구가 현재 30,000 명이고 매년 3%비율로 늘어나고 있다.

100,000 명의 인구가 될 때까지 얼마나 많은 해가 소요되는지를 계산하는

프로그램을 작성하시오.

```python
population = 30000       #인구수 할당
year = 0                 #년수 초기화

while population < 100000 :              #인구가 100000명이 되기전까지 반복
    population = population * (1 + 0.03) #매년 3%비율로 증가 연산
    year += 1                            #1번 루프마다 1년 증가
    
print("인구가 100,000명이 되기까지 ", year, "년이 소요되었습니다.")
```

문제설명 : 현재 인구가 30000명인 곳에 매년 3%의 비율로 인구가 늘어나고 있는데 100000명의

인구가 될 때까지 얼마나 많은 해가 소요되는지를 알아보기 위해 지나게 되는 년수를 입력받을 

year변수를 초기화하고 인구가 100000명이 넘지 않으면 계속 반복되는 while문을 사용하였습니다.

매년 3%비율로 증가함으로 population * (1 + 0.03)의 계산을 반복해서 해주게 되면 됩니다. 이 때

한 번 루프마다 year의 값을 1씩 증가시켜 몇해가 소요되는지를 알수 있습니다. 이렇게 계속 반복하여

루프가 돌면서 population의 값이 100000을 넘어가면 while문에서 나가지게되고 출력을 할 수 있습니다.

문제 4 번. (양수와 음수 개수 세기 및 평균 계산하기) 불특정 개수의 정수를 

읽은 후, 양수와 음수가 몇 개씩 읽혔는지 결정하고, 입력값의 개수와 평균을 

계산하는 프로그램을 작성하시오(0 은 세지 않는다). 프로그램은 입력값 0으로 

종료된다. 평균값을 부동소수점 숫자로 출력한다. 

실행 예는 다음과같다. 

정수를 입력하세요. 입력값이 0 이면 종료됩니다: 1 (enter)

정수를 입력하세요. 입력값이 0 이면 종료됩니다: 2 (enter)

정수를 입력하세요. 입력값이 0 이면 종료됩니다: -1 (enter)

정수를 입력하세요. 입력값이 0 이면 종료됩니다: 3 (enter)

정수를 입력하세요. 입력값이 0 이면 종료됩니다: 0 (enter)

양수의 개수는 3 개 입니다. 음수의 개수는 1 개 입니다.

총합은 5 입니다. 평균은 1.25 입니다

```python
num, plus, minus, count,sub = 0, 0, 0, 0, 0           #각 변수의 값을 초기화

while 1 :                                             #무한루프를 실행한다
    num = eval(input("정수를 입력하세요. 입력값이 0이면 종료됩니다. : "))
    if num > 0 :                                      #정수가 0보다 크다면
        plus += 1                                     #양수의 개수 값을 +1
    elif num < 0 :                                    #정수가 0보다 작다면
        minus += 1                                    #음수의 개수 값을 +1
    else :                                            #정수가 0이라면
        break                                         #무한루프를 종료해준다
    sub += num                                        #루프를 돌때마다 입력받은 정수를 합한다
    count += 1                                        #입력받은 정수 개수를 할당한다
    
print("양수의 개수는 ", plus, "개 입니다.")
print("음수의 개수는 ", minus, "개 입니다.")
print("총합은", sub, "입니다.")
print("평균은", sub / count, "입니다.")
        
```

문제설명 :정수를 입력받을 num변수, 그 값이 양수이면 +1받는 plus 변수, 값이 음수 이면 변수의 값을 +1받는 minus변수

sub는 입력받은 모든 정수들의 합을 입력받는 변수, count는 입력 받은 정수의 개수를 입력받습니다.

입력받은 정수가 0보다 크면 양수 이고 작으면 음수로 판별되어 양수면 plus에 +1 음수면 minus에 +1을 하여

입력받은 정수중에 양수가 몇개인지 음수가 몇개인지 판단할 수 있도록 하였습니다.

sub에 모든 정수의 값을 입력받아 count에 저장된 입력받은 정수의 개수로 나누어 주어 평균을 구하도록 하였습니다.

문제 5 번. (최대공약수 계산하기) 예제 코드 lec05/GreatestCommonDivisor.py에서

최대공약수를 구하는 방법을 알아보았다. 두 정수 n1 과 n2 의 최대공약수를 찾는 

또 다른 해결방법은 다음과 같다. 우선, n1 과 n2 중 작은 수를 d라고 한 후, d, 

d-1, d-2, …, 2, 1 의 순서로 각각 d 가 n1 과 n2 의 공약수인지 검사한다. 

첫 번째로 나타난 공약수가 두 수 n1 과 n2 의 최대공약수이다. 이 방법으로 

최대공약수를 구하는 프로그램을 작성하시오. 

```python
num1 = eval(input("첫 번째 정수를 입력하시오 : "))             #첫 번째 정수를 입력받는다
num2 = eval(input("두 번째 정수를 입력하시오 : "))             #두 번째 정수를 입력받는다

commonNum = 0                                                  #찾게 되는 최대 공약수를 넣을 변수

if num1 > num2 :                                               #처음 수가 크다면
    d = num2                                                   #나눌 수를 작은수로 정한다
    for i in range(0, d):
        if (num1 % (d-i) == 0 and num2 % (d-i) == 0) :        #첫 번째 숫자와 두 번째 숫자 모두 정햇던 작은 수로
            commonNum = d-i                                   #나누어 처음으로 나누어 떨어지게 한 수가 최대공약수
            break
    if (commonNum >= 2) :                                     #일반적으로 최대공약수의 범위 설정
        print("두 수의 최대공약수는 ", commonNum, "입니다.")
    else :
        print("두 수의 최대공약수는 없습니다.")
            
else :                                                        #위의 방법과 처음 수가 크냐 작냐의 차이
    d = num1                                                  #방법은 똑같다
    for i in range(0, d):
        if (num1 % (d-i) == 0 and num2 % (d-i) == 0) :
            commonNum = d-i
            break
    if (commonNum >= 2) :
        print("두 수의 최대공약수는 ", commonNum, "입니다.")
    else :
        print("두 수의 최대공약수는 없습니다.")
    
    
```

문제 설명 : 두 정수의 최대공약수를 찾기 위해서는 두 정수를 입력받습니다. 계산을 통해 나온 최대공약수 값을

할당하기 위해 commonNum이란 변수를 선언해줍니다. if문을 통해 두 수중 첫 번째 숫자가 크다면 작은 두 번째 

숫자를 나누는 수로 설정하고 그 숫자로 for문을 이용하여 값을 1씩 줄이면서 두 개의 숫자를 나누어 봅니다.

두 숫자를 모두 0으로 나누어 떨어지게 만드는 첫 번째 수가 두 숫자들의 최대공약수임을 알 수 있습니다.

아래의 코드는 위의 방법과 첫 번째 숫자가 크냐 작냐의 차이이지 방법은 똑같습니다.

문제 6 번. (윤년 출력하기) 21 세기(2001 년부터 2100 년까지)의 모든 윤년을 한 행에 10 개씩

출력하는 프로그램을 작성하시오. 연도는 단 공백 한 개로 구분된다.


```python
count = 0                                           #줄바꿈을 해주기위한 값을 받는 변수
for i in range (2001,2100) :                        #2001년 부터 2100년까지를 다 검사한다
    if ((i % 4) == 0 and (i % 100)) != 0 :          #4년마다 윤년이다. 100년 마다는 윤년이 아니다.
        print(i, end = " ")                         #윤년을 출력한다
        count += 1                                  #한행에 들어간 년도 개수를 증가
        if count == 10 :                            #한행에 10개가 들어가면 행을 바꿔준다
            print("\n")
            count = 0                               
        
```

문제 설명 : 21세기는 2001년 부터 2100년이기 때문에 윤년을 검사하는 범위는 2001년을 초기값,

2100을 종료값으로 설정합니다. 윤년은 4년마다 윤년이고 100년마다는 윤년이 아니며 400년마다는

윤년입니다. 하지만 문제에서 주어진 범위는 400년 기간을 포함하기 않기 때문에 400년은 주기는 

제외 시켰습니다. 4로 나누어 떨어지면서 100으로 나누어 떨이지지 않는 년수가 윤년임을 if문을

사용하여 설정하였고 한 한에 10개씩 출력을 하기 위해서 count변수를 사용하여 10개가 출력되면 

enter를 입력하도록 설정하여 코드를 완성시켰습니다.
