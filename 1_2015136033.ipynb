문제 1번. (정수의 자릿수 합산하기)) 0 과 1000 사이의 정수를 읽어서 각 자릿수의 합을 구하는 프로그램을 작성하시오.

예를 들어, 정수가 932 라면 자릿수의 합은 14 이다.(힌트: 자릿수를 추출하기 위해 % 연산자를 사용하고 추출된 자릿수를

없애기 위해 // 연산자를 사용하시오. 예를 들어, 932%10 = 2이고 932//10 = 93 이다.

ex) 0과 1000사이의 숫자를 입력하세요: 999(enter)

각 자릿수의 합은 27입니다.


```python
num = eval(input("0과 1000사이의 숫자를 입력하세요: "))

a = num//100      #백의 자릿수를 구해서 변수 a에 할당.
b = (num//10)%10  #십의 자릿수를 구해서 변수 b에 할당.
c = num%10        #일의 자릿수를 구해서 변수 c에 할당.

print("\n각 자릿수의 합은 " ,(a+b+c),"입니다.")
```

문제 설명 : 값을 입력받는 최대자릿수의 값이 100의 자릿수이다.

100의 자릿수는 //연산자를 이용하여 100을 나눠주게 되면 100의 자릿수가 나와 그값을 a변수에 할당합니다.

10의 자릿수는 입력받은 값에 //연산자를 이용해 10으로 나눠주게 되면 1의 자릿수를 제외한 값이 나오고 

그후 %연산자를 이용해 10으로 나눈후 나온 값을 변수 b에 할당합니다.

1의 자릿수는 입력 받은 값에 %연산자를 이용해 10으로 나눈후 나온값을 변수 c에 할당합니다.

문제 2번.(자동판매기 프로그램) 자동 판매기를 시뮬레이션하는 프로그램을 작성하여 보자. 

사용자는 1000 원짜리 지폐와 500 원짜리 동전, 100 원짜리 동전을 사용할 수 있다. 

물건값을 입력하고 1000 원권, 500 원짜리 동전, 100원짜리 동전의 개수를 입력하면 거스름돈을 계산하여서 동전으로 반환한다.

ex) 물건값을 입력하시오: 750 (enter)

     1000 원 지폐개수: 1
     
     500 원 동전개수: 0
     
     100 원 동전개수: 0
     
     500 원= 0  100 원= 2  10 원= 5  1 원=0

```python
money =  eval(input("물건의 값을 입력하시오: "))
money_1000 = eval(input("1000원 지폐개수: "))
money_500 = eval(input("500원 동전개수: "))
money_100 = eval(input("100원 동전개수: "))

first_value = ((money_1000*1000+money_500*500+money_100*100)-money)        #지불한 값들을 물건의 값과 계산한 것.

a = first_value//500                             #500원 동전 거스름돈 계산.
b = (first_value-500 * a)//100                   #100원 동전 거스름돈 계산.
c = (first_value-500 * a - 100 * b)//10          #10원 동전 거스름돈 계산.
d = first_value % 10                             #1원 동전 거스름돈 계산.

print("\n500원 = ", a ,"개 100원 = ", b ,"개 10원 = ", c , "개 1원 = ", d)
```

문제 설명 : 물건의 값을 받고 지불할 1000원 지폐개수, 500원 동전개수, 100원 동전개수를 입력합니다.

first_value라는 변수에 지불한 돈에서 물건의 값을 계산해주면 거스름돈의 값이 나오게 됩니다.

그 first_value의 값을 가지고 연산을 통해

500원 동전 거스름돈의 개수를 변수 a에 할당.

100원 동전 거스름돈의 개수를 변수 b에 할당.

10원 동전 거스름돈의 개수를 변수 c에 할당.

1원 동전 거스름돈의 개수를 변수 d에 할당.

따라서 거스름돈을 동전으로 환산하여 출력할 수 있습니다.

문제 3번.(년과 일수 계산하기) 분 값을 입력하고(예를 들어, 10 억분)

그 값에 대한 년수와 일수를 출력하는 프로그램을 작성하시오.

단순 계산을 위해, 1 년은 365 일이라고 가정한다.

ex)분에 대한 숫자를 입력하세요: 1000000000 (enter)
1000000000 분은 약 1902 년 214 일입니다.

```python
min = eval(input("분에 대한 숫자를 입력하세요: "))     #분을 입력받는다.

year = min//525600                   #1년은 525600분이다.
day = (min%525600)//1440             #하루는 1440분이다. 

print(min, "분은 약 ",year, "년 ", day, "일입니다.")
```

문제 설명 : min이라는 변수에 분을 입력받고 입력받은 분에 대한 년도를 계산하기 위해

525600분 = 1년 개념을 사용하여 입력받은 분을 나누어 년수를 구하고, 년수를 계산하고 

남은 나머지 분값에서 1일 = 1440분 개념을 사용하여 일수를 구해줄 수 있습니다.

문제 4번.(금융 애플리케이션: 급여) 다음의 정보를 읽고 급여 명세서를출력하는 프로그램을 작성하시오. 

사원의 이름(예, 정용제)
주당 근무 시간(예, 40)
시간당 급여(예, 9075)
원천징수세율(예, 20%)
주민세율(예, 9%)


ex)
사원이름을 입력하세요: 정용제 (enter)

주당 근무시간을 입력하세요: 40 (enter)

시간당 급여를 입력하세요: 9075 (enter)

원천징수세율을 입력하세요: 0.20 (enter)

지방세율을 입력하세요: 0.09 (enter)


사원 이름: 정용제

주당 근무시간: 40

임금: 9075

총 급여: 363000

공제:

     원천징수세 (20.0%): 72600
     주민세 (9.0%): 32670
     총 공제: 105270
     
공제 후 급여: 257730

```python
name = input("사원의 이름을 입력하세요: ")                  #이름을 입력받는다.
week_workhour = eval(input("주당 근무시간을 입력하세요: ")) #주당 근무시간을 입력받는다.
hour_money = eval(input("시간당 급여를 입력하세요: "))      #시간당 급여를 입력받는다.
tex = eval(input("원천징수세율을 입력하세요: "))            #원천징수세율을 입력받는다.
country_tex = eval(input("지방세율을 입력하세요: "))        #지방세율을 입력받는다.

all_money = week_workhour * hour_money                      #공제가 이루어지기전 총금액을 계산한다.
all_tex = (all_money * tex) + (all_money * country_tex)     #공제가 이루어질 금액을 계산한다.

print("\n사원 이름: ", name)
print("\n주당 근무시간: ", week_workhour)
print("\n임금: ", hour_money)
print("\n총 급여: ", all_money)
print("\n공제 : ")
print("\n  원천징수세 (",tex * 100,"%): ", all_money * tex)  #원천징수세율을 계산한 값을 출력
print("\n  주민세 (",country_tex * 100,"%): ", all_money * country_tex)   #지방세율을 계산한 값을 출력
print("\n  총 공제: ", all_tex)
print("\n공제 후 급여: ", all_money - all_tex)
```

문제 설명 : 각 해당하는 곳에 입력을 받고 이름, 주당, 임금, 총 급여, 공제 등을 출력해야합니다.

이름, 주당, 임금은 그대로 출력하여 주고 총 급여, 공제, 원청징수세, 주민세, 총 공제

공제 후 급여는 계산을 통해 출력해야합니다. 총 급여는 주당 근무시간 * 임금 계산하여 구하고

원청징수세는 (총 급여 * 원천징수세율)을 계산하여 구합니다. 주민세도 원천징수세와 같은

방법입니다. 총 공제값은 두 세금을 더한 값이고 공제 후 급여의 값은  총 공제금을 총 급여에서 

빼준값입니다. 이와 같은 과정을 통해 코드를 완성하였습니다.

문제 5 번. (금융 애플리케이션: 복지 가치) 연이율 5%의 계좌에 매달

100,000 원을 저축한다고 해보자. 연이율이 5%이므로 월이율은 0.05/12 = 0.00417 이다. 

첫 달 후, 계좌의 잔고는 다음과 같다. 100,000 * (1+0.00417) = 100,417

두 달 후, 계좌의 잔고는 다음과 같다. (100,000 + 100,417) * (1+0.00417) = 201,252

석 달 후, 계좌의 잔고는 다음과 같다. (100,000 + 201,252) * (1+0.00417) = 302,507

월저축액을 입력하고 6 개월 후에 계좌의 잔고를 화면에 출력하는 프로그램
을 작성하시오. 

다음은 프로그램의 실행 예이다. 월저축액: 100000 (enter)
6 개월 후 계좌의 잔고는 608811.35 입니다. 

```python
save_money = eval(input("월저축액 : "))    #월저축액을 입력받는다.
month_rate = 0.00417                       #월이율 할당.
i = 0                                      #for문을 돌리기 위한 변수 초기화.
monsave_money = 0                          #계산된 값을 밥는 변수 초기화.

for i in range(0,6):                       #6개월후 금액을 얻기위해 6번 반복 실행
    monsave_money = (monsave_money + save_money) * (1 + month_rate)
    #월이율이 추가되어 저축되는 연산이다.
    
print("6개월 후 계좌의 잔고는 ", format(monsave_money, "10.2f") , " 입니다.")
```

문제 설명 : 월저축액을 입력받고 월이율과의 계산을 해줍니다.

여기서 알고 싶은 값은 6개월 후의 계좌 잔고이기 때문에

for문을 사용하여 연산식을 6번 반복시켜 6개월이 지난 계좌 속 값을 만들

수 있습니다. for문안에 연산은 계좌속 금액이 한 달이 지나면 월이율이 붙고 

다음 달에 또 월저축액이 들어온값을 포함하여 또 월이율이 붙는 식으로 진행

되는 것을 위와 같은 소스코드 연산으로 표현하였습니다.

예와 같이 소숫점 2자리로 출력하기 위해 format()함수를 사용하였습니다.

문제 6 번. (기하학: 오각형의 넓이) 오각형의 넓이는 다음의 공식을 사용
하여 계산할 수 있다.(s 는 한 변의 길이이다.)

넓이 =(S*s^2)/(4*tan(pi/5))

사용자로부터 오각형 한 변의 길이를 입력받고 넓이를 출력하는 프로그램을작성하시오.

실행 예는 다음과 같다. 

한 변의 길이를 입력하세요: 5.5 (enter)

오각형의 넓이는 52.04444136781625 입니다

```python
import math                                         #tan()함수를 쓰기위해 math모듈을 import한다.
length = eval(input("한 변의 길이를 입력하세요 : "))
area = 0                                            #넓이 값을 받는 변수 초기화

area = (5*(length**2))/(4*math.tan(math.pi/5))     #오각형의 넓이를 구하는 연산식. 
                                                   #tan()함수와 제곱연산자**를 사용.
print("오각형의 넓이는 ", area , " 입니다.")
```

문제 설명 : 오각형의 연산식에 있어서 tan를 계산해야 하는데 tan()함수를 사용하기 위해

math모듈을 import하였습니다. 그 후 한 변의 길이를 입력받고 넓이를 구하는

공식을 이용해 tan()함수와 제곱연산자를 이용하여 오각형의 넓이를 구하여 출력하였습니다.
